# Recall@K Evaluation
Example notebook to evaluate recommendation quality.

```
def recall_at_k(actual, predicted, k):
    predicted = predicted[:k]
    return len(set(predicted) & set(actual)) / len(actual)
```

```
actual = ['Coffee','Bagel']
predicted = ['Coffee','Ice Capp','Donut','Wrap']
print('Recall@3:', recall_at_k(actual, predicted, 3))
```
